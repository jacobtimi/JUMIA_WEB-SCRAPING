```python
Descriptions = []
Ratings =  []
Price = []
for container in container:
    description = container.findAll("h3", {"class": "name"})
    for contain in container:
        dec = cont 
    descript = description[0].text
    Descriptions.append(descript)
        
    rating = container.findAll("div", {"class": "rev"})
    if len(rating)==0:
        Ratings.append(np.nan)
    else:
        RATE = rating[0].text
        Ratings.append(RATE)
    
    price = container.findAll("div", {"class": "prc"})
    price_product = price[0].text
    Price.append(price_product)
    
    print("description:"+ descript)
    print("rating:"+ RATE)
    print("price:"+ price_product)
```

```python
df = pd.DataFrame({"Product  Name":Descriptions, "Price":Price, "Rating": Ratings})
df.to_csv("Television_product.csv")
television = pd.read_csv("Television_product.csv")
television
```
